```python
import pandas as pd
df1 = pd.read_csv('countries1.csv')
print(df1)
```

```python

df3 = pd.read_csv('countries3.csv')
print(df3)
```

```python
# Combine rows from both DataFrames
df_vertical = pd.concat([df1, df3], axis=0)
print("Vertical concatenation:")
print(df_vertical)
```

```python
# Combine columns from both DataFrames
df_horizontal = pd.concat([df1, df3], axis=1)
print("Horizontal concatenation:")
print(df_horizontal)
```

```python
# Inner join (default) - only matching countries
countries_inner = pd.merge(df1, df3, on='Country', how='inner')

# Left join - all countries from df1, matching from df3
countries_left = pd.merge(df1, df3, on='Country', how='left')

# Right join - all countries from df3, matching from df1  
countries_right = pd.merge(df1, df3, on='Country', how='right')

# Outer join - all countries from both DataFrames
countries_outer = pd.merge(df1, df3, on='Country', how='outer')

print("Inner merge:")
print(countries_inner)

print("Left merge:")
print(countries_left)
```

```python
# Countries in either Eastern Europe OR Sub-Saharan Africa
filtered_or = df1[(df1['Region'] == 'ASIA') | (df1['Region'] == 'AFRICA')]
print("OR condition - Eastern Europe OR Sub-Saharan Africa:")
print(filtered_or)
```

```python
# Countries with population > 50 million AND GDP > 1000
filtered_and = df1[(df1['Population'] > 2000000) & (df1['Population'] > 1000)]
print("AND condition - Population > 50M AND GDP > 1000:")
print(filtered_and)
```

```python
import numpy as np

# Create sample arrays
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

print("Array 1:")
print(array1)
print("Shape:", array1.shape)

print("\nArray 2:")
print(array2)
print("Shape:", array2.shape)

# Concatenate on axis=0 (vertical - adds rows)
concat_axis0 = np.concatenate([array1, array2], axis=0)
print("\nConcatenate on axis=0 (vertical):")
print(concat_axis0)
print("Shape:", concat_axis0.shape)

# Concatenate on axis=1 (horizontal - adds columns)
concat_axis1 = np.concatenate([array1, array2], axis=1)
print("\nConcatenate on axis=1 (horizontal):")
print(concat_axis1)
print("Shape:", concat_axis1.shape)
```

```python
import pandas as pd
df = pd.read_csv('mushrooms.csv')
print(df.head(6))
```

```python
# If data is categorical, encode it
from sklearn.preprocessing import LabelEncoder
# Original data (categorical):
# edible: ['e', 'p', 'e', 'p']   [0, 1, 0, 1]
# cap-shape: ['x', 'b', 'x', 's']  [2, 0, 2, 1]
# cap-color: ['n', 'y', 'w', 'g']  [2, 3, 4, 1]
df_encoded = df.apply(LabelEncoder().fit_transform)

# Calculate and display correlation
correlation = df_encoded.corr()
print("Correlation Matrix:")
print(correlation)

```

```python
# Show correlation with target (first column)
print(f"\nCorrelations with target:")
print(correlation.iloc[0].sort_values(ascending=False))
```
