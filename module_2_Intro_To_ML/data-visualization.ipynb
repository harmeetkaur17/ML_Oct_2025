Key goals and components of EDA
Understand the data: EDA helps you get a deep understanding of what your data contains, including the types of variables, their distributions, and the overall structure. 
Uncover patterns and relationships: It reveals how different variables relate to each other, which is essential for building effective models. 
Identify data quality issues: You can find and address anomalies, missing values, or errors in the data that could impact your analysis. 
Generate hypotheses: EDA is a creative, open-ended process that allows you to form new questions and hypotheses based on initial findings.

```python
print("WHAT IS A BOX PLOT?")
print("="*20)

print("""
 A box plot displays:
    Median (middle line in box)
    First Quartile Q1 (25th percentile - bottom of box)
    Third Quartile Q3 (75th percentile - top of box)  
    Interquartile Range IQR = Q3 - Q1 (height of box)
    Whiskers (lines extending from box)
    Outliers (points beyond whiskers)

ANATOMY OF A BOX PLOT:
    
    *   Outlier
    |
      Upper whisker (Q3 + 1.5IQR)
    
   
      Q3 (75th percentile)
     Median (50th percentile)  
      Q1 (25th percentile)
   
    
      Lower whisker (Q1 - 1.5IQR)
    |
    *   Outlier
""")
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("\n1. BASIC BOX PLOT EXAMPLE:")
print("="*30)

# Sample data - test scores
test_scores = [45, 52, 58, 63, 68, 72, 75, 78, 82, 85, 88, 92, 98, 105]

print("Test scores data:", test_scores)

# Calculate key statistics
q1 = np.percentile(test_scores, 25)
median = np.percentile(test_scores, 50) 
q3 = np.percentile(test_scores, 75)
iqr = q3 - q1

print(f"\nKey Statistics:")
print(f"Q1 (25th percentile): {q1}")
print(f"Median (50th percentile): {median}")
print(f"Q3 (75th percentile): {q3}")
print(f"IQR (Q3 - Q1): {iqr}")

# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot(test_scores, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', alpha=0.7))
plt.ylabel('Test Scores')
plt.title('Distribution of Test Scores')
plt.grid(True, alpha=0.3)
plt.show()

print(" The box shows the middle 50% of data (Q1 to Q3)")
print(" The whiskers extend to the furthest points within 1.5IQR")
```

```python
print("1. OUTLIER DETECTION WITH BOX PLOTS")
print("="*40)

import matplotlib.pyplot as plt
import numpy as np

print(" HOW BOX PLOTS DETECT OUTLIERS:")
print("""
 Outlier Detection Rule:
    Calculate IQR = Q3 - Q1
    Lower fence = Q1 - 1.5  IQR
    Upper fence = Q3 + 1.5  IQR
    Any point outside these fences = OUTLIER

 Visual Identification:
    Points beyond whiskers = Outliers
    Shown as individual dots/circles
    Easy to spot at a glance
""")

# Example with outliers
salary_data = [35000, 38000, 40000, 42000, 45000, 47000, 50000, 52000, 
               55000, 58000, 60000, 62000, 65000, 200000, 250000]  

print("Sample salary data with outliers:")
print(f"Data: {salary_data}")

# Calculate outlier boundaries
q1 = np.percentile(salary_data, 25)
q3 = np.percentile(salary_data, 75)
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

print(f"\nOutlier Detection Calculation:")
print(f"Q1 (25th percentile): ${q1:,.0f}")
print(f"Q3 (75th percentile): ${q3:,.0f}")
print(f"IQR: ${iqr:,.0f}")
print(f"Lower fence: ${lower_fence:,.0f}")
print(f"Upper fence: ${upper_fence:,.0f}")

# Find outliers
outliers = [x for x in salary_data if x < lower_fence or x > upper_fence]
normal_values = [x for x in salary_data if lower_fence <= x <= upper_fence]

print(f"\n OUTLIERS DETECTED: {outliers}")
print(f" Normal values: {len(normal_values)} out of {len(salary_data)}")

# Create box plot
plt.figure(figsize=(12, 8))
box_plot = plt.boxplot(salary_data, patch_artist=True,
                       boxprops=dict(facecolor='lightblue', alpha=0.7),
                       flierprops=dict(marker='o', markerfacecolor='red', 
                                     markersize=10, alpha=0.8))

plt.ylabel('Salary ($)')
plt.title('Salary Distribution - Outlier Detection')
plt.grid(True, alpha=0.3)


plt.show()

print(" Red dots beyond whiskers = OUTLIERS")
print(" Whiskers extend to farthest non-outlier points")
```

```python
print("WHAT ARE VIOLIN PLOTS?")
print("="*25)

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

print("""
 VIOLIN PLOT = Box Plot + Density Plot

 A violin plot shows:
    Distribution density (width of violin shape)
    All quartiles (Q1, Median, Q3)
    Data spread and shape
    Multiple peaks (multimodal distributions)
    Skewness direction and degree

 Think of it as:
   Box Plot: Shows SUMMARY statistics only
   Violin Plot: Shows COMPLETE distribution shape

VIOLIN PLOT ANATOMY:
    
     
       
        Median line
         Box plot elements inside
        
            Wider = More data density
        
         Q1 and Q3 lines
   ___    Narrower = Less data density
""")
```

```python
print("\n1. BASIC VIOLIN PLOT EXAMPLE")
print("="*35)

# Create sample data
np.random.seed(42)
normal_data = np.random.normal(50, 10, 200)

print("Sample data: 200 points from normal distribution")
print(f"Mean: {np.mean(normal_data):.1f}")
print(f"Standard deviation: {np.std(normal_data):.1f}")

# Create violin plot
plt.figure(figsize=(12, 8))
violin_plot = plt.violinplot(normal_data, showmeans=True, showmedians=True, showextrema=True)

# Customize appearance
for pc in violin_plot['bodies']:
    pc.set_facecolor('lightblue')
    pc.set_alpha(0.7)

plt.ylabel('Values')
plt.title('Basic Violin Plot - Normal Distribution')
plt.grid(True, alpha=0.3)
plt.show()

print(" VIOLIN PLOT COMPONENTS:")
print(" Wide parts = High data density")
print(" Narrow parts = Low data density") 
print(" Inner lines = Quartiles and median")
print(" Symmetrical shape = Normal distribution")
```

```python
print("WHAT ARE HEAT MAPS?")
print("="*20)

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

print("""
 HEAT MAP = Data visualization using colors to represent values

 A heat map shows:
    Data values as colors (dark/light, hot/cold colors)
    Missing data patterns

HEAT MAP COMPONENTS:
    
   High Values   Red/Hot colors
   Medium Values   Yellow/Warm colors  
   Low Values   Blue/Cold colors
   Missing Values   White/Gray
""")
```

```python
print("\n1. BASIC HEAT MAP EXAMPLE")
print("="*30)

# Create sample data matrix
np.random.seed(42)
data_matrix = np.random.randn(8, 6) * 10 + 50

# Create labels
rows = ['Product A', 'Product B', 'Product C', 'Product D', 
        'Product E', 'Product F', 'Product G', 'Product H']
cols = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# Convert to DataFrame for better labeling
df = pd.DataFrame(data_matrix, index=rows, columns=cols)

print("Sample sales data (in thousands):")
print(df.round(1))

# Create basic heat map
plt.figure(figsize=(10, 8))
plt.imshow(data_matrix, cmap='YlOrRd', aspect='auto')
plt.colorbar(label='Sales (in thousands)')
plt.title('Monthly Sales Heat Map - Basic')

# Add labels
plt.xticks(range(len(cols)), cols)
plt.yticks(range(len(rows)), rows)


plt.show()

print(" HEAT MAP INSIGHTS:")
print(" Red/Orange = High sales")
print(" Yellow = Medium sales")
print(" Lighter colors = Lower sales")
print(" Easy to spot patterns across months and products")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the iris dataset
df = sns.load_dataset("iris")

# Method 1: Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[float, int])

# Create a correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), 
            annot=True,           # Show correlation values
            cmap='coolwarm',      # Color scheme
            center=0,             # Center colormap at 0
            square=True,          # Make cells square
            fmt='.2f',            # Format to 2 decimal places
            cbar_kws={'label': 'Correlation'})

plt.title('Iris Dataset - Correlation Heatmap (Numeric Features Only)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the iris dataset
df = sns.load_dataset("iris")

# Create heatmap showing average values by species
plt.figure(figsize=(10, 6))
species_means = df.groupby('species').mean()

sns.heatmap(species_means, 
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            cbar_kws={'label': 'Average Value (cm)'})

plt.title('Iris Dataset - Average Feature Values by Species')
plt.ylabel('Species')
plt.xlabel('Features')
plt.show()
```
