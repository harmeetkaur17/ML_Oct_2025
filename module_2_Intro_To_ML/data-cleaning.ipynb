```python
def update_mfr(val):
    mapping = {
        'A': 'American Home Food Products',
        'G': 'General Mills',
        'K': 'Kelloggs',
        'N': 'Nabisco',
        'P': 'Post',
        'Q': 'Quaker Oats',
        'R': 'Ralston Purina'
    }
    manufacturer_full_name = mapping[val]
    return manufacturer_full_name


df['mfr'] = df['mfr'].apply(update_mfr)
df.head(10)
```
