```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
len(df)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
# Calculate accuracy score using sklearn
from sklearn.metrics import accuracy_score

# Method 1: Using sklearn accuracy_score function
accuracy = accuracy_score(y_test, y_pred)

print("ðŸŽ¯ ACCURACY SCORE CALCULATION")
print("="*35)
print(f"Sklearn Accuracy Score: {accuracy:.4f}")
print(f"Accuracy Percentage: {accuracy:.2%}")
```

```python
prediction = model.predict([[4.8,3.0,1.5,0.3]])
print(f"Prediction for a new flower: {prediction[0]}")
```

```python

# Convert prediction number to flower name
flower_names = iris.target_names
predicted_flower = flower_names[prediction[0]] # flower_name[0]
print(f"The flower type is: {predicted_flower}")
```

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```

```python
print(f"\nðŸŽ¯ WHEN TO USE LINEAR vs NON-LINEAR SVM")
print("="*45)

print("ðŸ“ USE LINEAR SVM WHEN:")
print("â€¢ Data appears linearly separable")
print("â€¢ High-dimensional data (text classification)")
print("â€¢ Large datasets (faster training)")
print("â€¢ Interpretability is important")
print("â€¢ Limited computational resources")
print("â€¢ Example: Spam detection, document classification")

print(f"\nðŸŒ€ USE NON-LINEAR SVM WHEN:")
print("â€¢ Data has complex patterns")
print("â€¢ Classes overlap significantly")
print("â€¢ Small to medium-sized datasets")
print("â€¢ Higher accuracy is prioritized over speed")
print("â€¢ Feature relationships are complex")
print("â€¢ Example: Image classification, pattern recognition")

print(f"\nðŸ’¡ YOUR IRIS RESULTS INTERPRETATION:")
linear_score = SVC(kernel='linear').fit(X_train, y_train).score(X_test, y_test)
rbf_score = SVC(kernel='rbf').fit(X_train, y_train).score(X_test, y_test)

print(f"Linear SVM accuracy: {linear_score:.1%}")
print(f"RBF (non-linear) SVM accuracy: {rbf_score:.1%}")

if linear_score >= rbf_score:
    print("â†’ Linear kernel performs as well as non-linear")
    print("â†’ Iris data has good linear separability")
    print("â†’ Simpler model (linear) is preferred")
else:
    print("â†’ Non-linear kernel performs better")
    print("â†’ Iris data benefits from complex boundaries")
    print("â†’ Non-linear relationships exist in the data")
```
