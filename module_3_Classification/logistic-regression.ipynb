```python
import pandas as pd
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Predict for 5 new people with different ages
new_people = [[22], [28], [35], [42], [55]]


print("üéØ INSURANCE PREDICTIONS FOR 5 NEW PEOPLE")
print("="*50)

for people in new_people:
    # Get binary prediction (0 or 1)
    prediction = model.predict([people])

    
    # Interpret result
    will_buy = "Yes" if prediction[0] == 1 else "No"

    print(f"  Age: {people[0]}")
    print(f"  Prediction: {prediction[0]} ({will_buy})")
   
```

```python
# Import required libraries for confusion matrix
from sklearn.metrics import confusion_matrix

print("üìä CONFUSION MATRIX FOR INSURANCE PREDICTION MODEL")
print("="*55)
```

```python
# Create confusion matrix using test predictions
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

print("üî¢ CONFUSION MATRIX (Raw Numbers):")
print(cm)

# Extract the four components
tn, fp, fn, tp = cm.ravel()

print(f"\nüìã CONFUSION MATRIX COMPONENTS:")
print(f"True Negatives (TN):  {tn} - Correctly predicted 'Will NOT buy insurance'")
print(f"False Positives (FP): {fp} - Wrongly predicted 'Will buy insurance' (False Alarm)")
print(f"False Negatives (FN): {fn} - Wrongly predicted 'Will NOT buy' (Missed Sale)")
print(f"True Positives (TP):  {tp} - Correctly predicted 'Will buy insurance'")

print(f"\nüìä MATRIX LAYOUT:")
print("                 PREDICTED")
print("            Will NOT Buy  Will Buy")
print("         ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê")
print(f"ACTUAL   ‚îÇ      {tn:2d}         {fp:2d}    ‚îÇ Will NOT Buy")
print(f"Will Buy ‚îÇ      {fn:2d}         {tp:2d}    ‚îÇ Will Buy")
print("         ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò")
```

```python
# Calculate accuracy score using sklearn
from sklearn.metrics import accuracy_score

# Method 1: Using sklearn accuracy_score function
accuracy = accuracy_score(y_test, y_pred)

print("üéØ ACCURACY SCORE CALCULATION")
print("="*35)
print(f"Sklearn Accuracy Score: {accuracy:.4f}")
print(f"Accuracy Percentage: {accuracy:.2%}")

# 5 / 6 * 100 = 83.33%
```

```python
# Calculate precision score using sklearn
from sklearn.metrics import precision_score

# Method 1: Using sklearn precision_score function
precision = precision_score(y_test, y_pred)

print("üéØ PRECISION SCORE CALCULATION")
print("="*35)
print(f"Sklearn Precision Score: {precision:.4f}")
print(f"Precision Percentage: {precision:.2%}")

# 2 / 3 * 100 = 66.67%
```

```python
# Calculate recall score using sklearn
from sklearn.metrics import recall_score

# Method 1: Using sklearn recall_score function
recall = recall_score(y_test, y_pred)

print("üéØ RECALL SCORE CALCULATION")
print("="*32)
print(f"Sklearn Recall Score: {recall:.4f}")
print(f"Recall Percentage: {recall:.2%}")

# 2 / 2 = 100%
```
