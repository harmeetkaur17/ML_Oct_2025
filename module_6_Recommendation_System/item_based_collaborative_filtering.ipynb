```python
import os
import numpy as np
import pandas as pd
```

```python
ratings_data = pd.read_csv('ratings.csv')
ratings_data.head()
```

```python
movie_data = pd.read_csv('movies.csv')
movie_data.head()
```

```python
movie_rating_data = pd.merge(ratings_data, movie_data, on='movieId')
movie_rating_data.head(5)
```

```python
movie_rating_data['rating'] = movie_rating_data.groupby('movieId')['rating'].mean()
movie_rating_data.head(5)
```

```python
#Create user-iten matrix
user_movie_matrix = movie_rating_data.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
)
user_movie_matrix
```

```python
user_movie_matrix_filled = user_movie_matrix.fillna(0)
user_movie_matrix_filled
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_similarity = cosine_similarity(user_movie_matrix_filled.T)
item_similarity
```

```python
import pandas as pd

item_similarity_df = pd.DataFrame(
    item_similarity,
    index=user_movie_matrix.columns,
    columns=user_movie_matrix.columns
)
item_similarity_df
```

```python
user_ratings = user_movie_matrix_filled.loc[1]
unrated_movies = user_ratings[user_ratings == 0].index  # movies not rated by user
unrated_movies
```

```python

 
predicted_ratings = {}
    
for movie in unrated_movies:
    # Step 4a: Get similarity scores of this movie with all other movies
    sim_scores = item_similarity_df[movie]
    
    # Step 4b: Keep only movies that the user has actually rated
    rated_movies = user_ratings[user_ratings > 0].index
    sim_scores = sim_scores[rated_movies]
    
    # Step 4c: Compute predicted rating using weighted average
    if sim_scores.sum() > 0:
        predicted_rating = np.dot(sim_scores, user_ratings[rated_movies]) / sim_scores.sum()
        predicted_ratings[movie] = predicted_rating

predicted_ratings

```

```python
recommended_movies = sorted(predicted_ratings.items(), key=lambda x: x[1], reverse=True)[:10]
recommended_movies

```

```python
recommended_movies_ids = [movie[0] for movie in recommended_movies]
recommended_movies_ids
```

```python
# Create a DataFrame of recommended movie IDs
recommended_df = pd.DataFrame(recommended_movies_ids, columns=['movieId'])

# Merge with movie_data to get titles
top_recommendations = recommended_df.merge(movie_data[['movieId', 'title']], on='movieId', how='left')

print(top_recommendations)
```
