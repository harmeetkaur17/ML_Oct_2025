```python
import os
import numpy as np
import pandas as pd
```

```python
ratings_data = pd.read_csv('ratings.csv')
ratings_data.head()
```

```python
movie_data = pd.read_csv('movies.csv')
movie_data.head()
```

```python
movie_rating_data = pd.merge(ratings_data, movie_data, on='movieId')
```

```python
movie_rating_data.head(5)
```

```python
#Create user-iten matrix
user_movie_matrix = movie_rating_data.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
)
user_movie_matrix
```

```python
user_movie_matrix_filled = user_movie_matrix.fillna(0)
user_movie_matrix_filled
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_movie_matrix_filled)
user_similarity
```

```python
import pandas as pd

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)
user_similarity_df
```

```python
similar_users = user_similarity_df[1].sort_values(ascending=False)
similar_users
```

```python
similar_users = similar_users.drop(1)
similar_users
```

```python
target_user_movies = user_movie_matrix.loc[1]
target_user_movies
```

```python
unwatched_movies = target_user_movies[target_user_movies.isna()].index
unwatched_movies
```

```python
predicted_ratings = {}

for movie in unwatched_movies:
    numerator = 0
    denominator = 0

    for similar_user, similarity in similar_users.items():
        rating = user_movie_matrix.loc[similar_user, movie]

        if not pd.isna(rating):
            numerator += similarity * rating
            denominator += similarity

    if denominator != 0:
        predicted_ratings[movie] = numerator / denominator
```

```python
predicted_ratings = pd.Series(predicted_ratings).sort_values(ascending=False)
predicted_ratings
```

```python
# Get top 10 predicted movies
top_10_movies = predicted_ratings.head(10)

# Map movieIds to titles
top_10_with_titles = top_10_movies.reset_index()
top_10_with_titles.columns = ['movieId', 'predicted_rating']
top_10_with_titles = top_10_with_titles.merge(movie_data[['movieId', 'title']], on='movieId')

print("Top 10 recommended movies for user 1:")
top_10_with_titles
```
