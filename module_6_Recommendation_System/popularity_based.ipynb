```python
import os
import numpy as np
import pandas as pd
```

```python
ratings_data = pd.read_csv('ratings.csv')
ratings_data.head()
```

```python
movie_data = pd.read_csv('movies.csv')
movie_data.head()
```

```python
movie_rating_data = pd.merge(ratings_data, movie_data, on='movieId')
```

```python
movie_rating_data.head(30)
```

```python
movie_rating_data.groupby('title')['rating'].mean().head()
```

```python
movie_rating_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
movie_rating_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating_mean_count = pd.DataFrame(movie_rating_data.groupby('title')['rating'].mean())
rating_mean_count['rating_counts'] = pd.DataFrame(movie_rating_data.groupby('title')['rating'].count())
rating_mean_count.head()
```

```python
rating_mean_count[rating_mean_count['rating'] == 5.0].sort_values(by='rating', ascending=False).head(5)
```

```python
# Get top 5 movies with at least 50 ratings, sorted by average rating
rating_mean_count[rating_mean_count['rating_counts'] >= 50].sort_values(by='rating', ascending=False).head(5)
```

```python
C = rating_mean_count['rating'].mean()
m = rating_mean_count['rating_counts'].quantile(0.90)

def imdb_score(row):
    v = row['rating_counts']
    R = row['rating']
    return (v/(v+m) * R) + (m/(v+m) * C)

rating_mean_count['imdb_score'] = rating_mean_count.apply(imdb_score, axis=1)
```

```python
rating_mean_count.sort_values('imdb_score', ascending=False).head(5)
```
